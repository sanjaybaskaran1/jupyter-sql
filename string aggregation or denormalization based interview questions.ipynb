```python
%%sql
with unique_name_$_resources as
(
select name, resources
from users 
group by name, resources
)
select name, group_concat(resources, ' ') as "Resources"
from unique_name_$_resources
group by name;
```

```python
%%sql
with unique_name_$_resources as
(
select name, resources
from users 
group by name, resources
)
, cte2 as
(
select name, group_concat(resources, ' ') as "Resources"
from unique_name_$_resources
group by name
)
select t2.name, t2.Resources,count(u.floor) as "total_floor_visits"
from users as `u` inner join cte2 as `t2`
on u.name = t2.name
group by u.name;
```

```python
%%sql
with unique_name_$_resources as
(
select name, resources
from users 
group by name, resources
)
, cte2 as
(
select name, group_concat(resources, ' ') as "Resources"
from unique_name_$_resources
group by name
)
, cte3 as
(
select t2.name, t2.Resources,count(u.floor) as "total_floor_visits"
from users as `u` inner join cte2 as `t2`
on u.name = t2.name
group by u.name
)
select t2.name,t2.resources, t2.total_floor_visits, u.floor
from users as u inner join cte3 as t2
on u.name = t2.name;
```

```python
%%sql
with unique_name_$_resources as
(
select name, resources
from users 
group by name, resources
)
, cte2 as
(
select name, group_concat(resources, ' ') as "Resources"
from unique_name_$_resources
group by name
)
, cte3 as
(
select t2.name, t2.Resources,count(u.floor) as "total_floor_visits"
from users as `u` inner join cte2 as `t2`
on u.name = t2.name
group by u.name
)
, cte4 as
(
select t2.name,t2.resources, t2.total_floor_visits, u.floor
from users as u inner join cte3 as t2
on u.name = t2.name
)
select name,resources,total_floor_visits,floor as "most_visited_floor"
from cte4
group by name,resources,total_floor_visits,floor
having count(floor) > 1;
```
