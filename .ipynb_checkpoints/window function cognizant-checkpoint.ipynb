```python
%%sql
  SELECT
        student_name,
        SUBSTRING_INDEX(mark_list, ',', 1) AS mark,
        SUBSTRING(mark_list, LENGTH(SUBSTRING_INDEX(mark_list, ',',1))+2) AS rest
    FROM students_new;
```

```python
%%sql
WITH RECURSIVE split_marks AS (
    -- Anchor member: Get the first mark and the remaining marks
    SELECT
        student_name,
        SUBSTRING_INDEX(mark_list, ',', 1) AS mark,
        SUBSTRING(mark_list, LENGTH(SUBSTRING_INDEX(mark_list, ',', 1)) + 2) AS rest
    FROM students_new

    UNION ALL

    -- Recursive member: Keep extracting marks until rest is empty
    SELECT
        student_name,
        SUBSTRING_INDEX(rest, ',', 1) AS mark,
        SUBSTRING(rest, LENGTH(SUBSTRING_INDEX(rest, ',', 1)) + 2) AS rest
    FROM split_marks
    WHERE rest != ''
)

-- Final selection of student names and their individual marks
SELECT student_name, mark, rest
FROM split_marks
WHERE mark != '';
```

```python
%%sql
WITH RECURSIVE split_marks AS (
    -- Anchor member: Get the first mark and the remaining marks
    SELECT
        student_name,
        SUBSTRING_INDEX(mark_list, ',', 1) AS mark,
        SUBSTRING(mark_list, LENGTH(SUBSTRING_INDEX(mark_list, ',', 1)) + 2) AS rest
    FROM students_new

    UNION ALL

    -- Recursive member: Keep extracting marks until rest is empty
    SELECT
        student_name,
        SUBSTRING_INDEX(rest, ',', 1) AS mark,
        SUBSTRING(rest, LENGTH(SUBSTRING_INDEX(rest, ',', 1)) + 2) AS rest
    FROM split_marks
    WHERE rest != ''
)
, `subject row no` AS (
    SELECT student_name, mark, 
           ROW_NUMBER() OVER (PARTITION BY student_name) AS `row num`
    FROM split_marks
)
, `subject rank` AS (
    SELECT student_name, mark, `row num`,
           RANK() OVER (PARTITION BY `row num` ORDER BY mark DESC) AS `rank`
    FROM `subject row no`
)
SELECT * FROM `subject rank`;
```

```python
%%sql
WITH RECURSIVE split_marks AS (
    -- Anchor member: Get the first mark and the remaining marks
    SELECT
        student_name,
        SUBSTRING_INDEX(mark_list, ',', 1) AS mark,
        SUBSTRING(mark_list, LENGTH(SUBSTRING_INDEX(mark_list, ',', 1)) + 2) AS rest
    FROM students_new

    UNION ALL

    -- Recursive member: Keep extracting marks until rest is empty
    SELECT
        student_name,
        SUBSTRING_INDEX(rest, ',', 1) AS mark,
        SUBSTRING(rest, LENGTH(SUBSTRING_INDEX(rest, ',', 1)) + 2) AS rest
    FROM split_marks
    WHERE rest != ''
)
, `subject row no` AS (
    SELECT student_name, mark, 
           ROW_NUMBER() OVER (PARTITION BY student_name) AS `row num`
    FROM split_marks
)
, `subject rank` AS (
    SELECT student_name, mark, `row num`,
           RANK() OVER (PARTITION BY `row num` ORDER BY mark DESC) AS `rank`
    FROM `subject row no`
)
SELECT * FROM `subject rank`
where `rank` in(1);
```

```python
%%sql
select unnest(string_to_array(mark_list, ',')) from students_new;
```

```python
%%sql
select student_name, 
unnest(string_to_array(mark_list, ',')) as "marks"
from students_new;
```

```python
%%sql
with students_marks as
(
select student_name, 
unnest(string_to_array(mark_list, ',')) as "marks"
from students_new
)
select student_name,marks,
row_number() over(partition by student_name) as "subjects"
from students_marks;
```

```python
%%sql

with students_marks as
(
select student_name, 
unnest(string_to_array(mark_list, ',')) as "marks"
from students_new
)

,students_marks_subject_wise as
(
select student_name,marks,
row_number() over(partition by student_name) as "subjects"
from students_marks
)

select student_name,marks,subjects,
dense_rank() over(partition by subjects order by marks desc) as "subject_ranks"
from students_marks_subject_wise;
```

--> find the students who got top-2 ranks in each subjects

```python
%%sql

with students_marks as
(
select student_name, 
unnest(string_to_array(mark_list, ',')) as "marks"
from students_new
)

,students_marks_subject_wise as
(
select student_name,marks,
row_number() over(partition by student_name) as "subjects"
from students_marks
)

,subject_wise_marks as
(
select student_name,marks,subjects,
dense_rank() over(partition by subjects order by marks desc) as "subject ranks"
from students_marks_subject_wise
)

select * from subject_wise_marks
where "subject ranks" in(1,2);
```
