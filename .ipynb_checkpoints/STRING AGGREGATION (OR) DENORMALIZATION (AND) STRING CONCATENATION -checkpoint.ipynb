```python
%%sql
select * from users;
```

```python
%%sql
select name, resources
from users 
group by name, resources;
```

```python
%%sql
with unique_name_$_resources as
(
select name, resources
from users 
group by name, resources
)
select name, group_concat(resources, ' ') as "Resources"
from unique_name_$_resources
group by name;
```

```python
%%sql
with unique_name_$_resources as
(
select name, resources
from users 
group by name, resources
)
, cte2 as
(
select name, group_concat(resources, ' ') as "Resources"
from unique_name_$_resources
group by name
)
select t2.name, t2.Resources,count(u.floor) as "total_floor_visits"
from users as `u` inner join cte2 as `t2`
on u.name = t2.name
group by u.name;
```

```python
%%sql
with unique_name_$_resources as
(
select name, resources
from users 
group by name, resources
)
, cte2 as
(
select name, group_concat(resources, ' ') as "Resources"
from unique_name_$_resources
group by name
)
, cte3 as
(
select t2.name, t2.Resources,count(u.floor) as "total_floor_visits"
from users as `u` inner join cte2 as `t2`
on u.name = t2.name
group by u.name
)
select t2.name,t2.resources, t2.total_floor_visits, u.floor
from users as u inner join cte3 as t2
on u.name = t2.name;
```

```python
%%sql
with unique_name_$_resources as
(
select name, resources
from users 
group by name, resources
)
, cte2 as
(
select name, group_concat(resources, ' ') as "Resources"
from unique_name_$_resources
group by name
)
, cte3 as
(
select t2.name, t2.Resources,count(u.floor) as "total_floor_visits"
from users as `u` inner join cte2 as `t2`
on u.name = t2.name
group by u.name
)
, cte4 as
(
select t2.name,t2.resources, t2.total_floor_visits, u.floor
from users as u inner join cte3 as t2
on u.name = t2.name
)
select name,resources,total_floor_visits,floor as "most_visited_floor"
from cte4
group by name,resources,total_floor_visits,floor
having count(floor) > 1;
```

```python
%%sql
select * from Combo;
```

```python
%%sql
select *,
unnest(string_to_array(combo, ',')) as "split_combo"
from Combo;
```

```python
%%sql
with split_string as
(
select combo, cost,
unnest(string_to_array(combo, ',')) as "split_combo"
from combo
)
select combo,cost,
string_agg(split_combo, ','  ORDER BY split_combo ASC) as "merged_combo"
from split_string
group by combo, cost;
```

```python
%%sql
with split_string as
(
select combo, cost,
unnest(string_to_array(combo, ',')) as "split_combo"
from combo
)
, merged_string as
(
select combo,cost,
string_agg(split_combo, ','  ORDER BY split_combo ASC) as "merged_combo"
from split_string
group by combo, cost
)
select * 
from merged_string
where combo = merged_combo;
```

```python
%%sql
select *
from details;
```

```python
%%sql
SELECT col1,col2,col3,
       STRING_AGG(CONCAT(col1, ',' ,col2, ','  ,col3), ',' ORDER BY col2, col3) AS concatenated_cols
FROM details
GROUP BY col1,col2,col3;
```

```python
%%sql
with cte as
(
SELECT col1,col2,col3,
STRING_AGG(CONCAT(col1, ',' ,col2, ','  ,col3), ',' ORDER BY col2, col3) AS concatenated_cols
FROM details
GROUP BY col1,col2,col3
)
select *
from cte inner join details d
on cte.col1=d.col1 and cte.col2=d.col2 and cte.col3=d.col3;
```

```python
%%sql
with cte as
(
SELECT col1,col2,col3,
STRING_AGG(CONCAT(col1, ',' ,col2, ','  ,col3), ',' ORDER BY col2, col3) AS concatenated_cols
FROM details
GROUP BY col1,col2,col3
)
, cte2 as
(
select *
from cte inner join details d
on cte.col1=d.col1 and cte.col2=d.col2 and cte.col3=d.col3
)
select concatenated_cols,
unnest(string_to_array(concatenated_cols, ',')) as "split_combo",
score
from cte2
```

```python
%%sql
with cte as
(
SELECT col1,col2,col3,
STRING_AGG(CONCAT(col1, ',' ,col2, ','  ,col3), ',' ORDER BY col2, col3) AS concatenated_cols
FROM details
GROUP BY col1,col2,col3
)
, cte2 as
(
select *
from cte inner join details d
on cte.col1=d.col1 and cte.col2=d.col2 and cte.col3=d.col3
)
, cte3 as
(
select concatenated_cols,
unnest(string_to_array(concatenated_cols, ',')) as "split_combo",
score
from cte2
)
select concatenated_cols, score,
string_agg(split_combo, ',' order by split_combo asc) as "agg_column"
from cte3
group by concatenated_cols, score;
```

```python
%%sql
with cte as
(
SELECT col1,col2,col3,
STRING_AGG(CONCAT(col1, ',' ,col2, ','  ,col3), ',' ORDER BY col2, col3) AS concatenated_cols
FROM details
GROUP BY col1,col2,col3
)
, cte2 as
(
select *
from cte inner join details d
on cte.col1=d.col1 and cte.col2=d.col2 and cte.col3=d.col3
)
, cte3 as
(
select concatenated_cols,
unnest(string_to_array(concatenated_cols, ',')) as "split_combo",
score
from cte2
)
, cte4 as
(
select concatenated_cols, score,
string_agg(split_combo, ',' order by split_combo asc) as "agg_column"
from cte3
group by concatenated_cols, score
)
select agg_column, sum(score) as "Total_score_by_combination"
from cte4
group by agg_column
order by agg_column asc;
```
