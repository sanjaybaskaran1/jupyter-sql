## find the last 3 months average sales amount for each customers {TCS-INTERVIEW}

```python
%%sql
select * from CustomerPurchases;
```

```python
%%sql
SELECT Name,PurchaseDate,MONTH(PurchaseDate) AS "PurchaseMonth", year(PurchaseDate) as "PurchseYear",
Amount
FROM CustomerPurchases;
```

```python
%%sql
SELECT Name,MONTH(PurchaseDate) AS "PurchaseMonth", year(PurchaseDate) as "PurchseYear",
sum(Amount) as "TotalAmount"
FROM CustomerPurchases
group by Name,PurchaseMonth,PurchseYear;
```

```python
%%sql
with cte as
(
SELECT Name,MONTH(PurchaseDate) AS "PurchaseMonth", year(PurchaseDate) as "PurchaseYear",
sum(Amount) as "TotalAmount"
FROM CustomerPurchases
group by Name,PurchaseMonth,PurchaseYear
)
select Name, PurchaseMonth,TotalAmount,
avg(TotalAmount) over(partition by Name order by PurchaseYear,PurchaseMonth 
                                      rows between 2 preceding and current row ) as "3-MonthsAvgAmount"
from cte;
```
