```python
%%sql
WITH RECURSIVE split_marks AS (
    -- Anchor member: Get the first mark and the remaining marks
    SELECT
        student_name,
        SUBSTRING_INDEX(mark_list, ',', 1) AS mark,
        SUBSTRING(mark_list, LENGTH(SUBSTRING_INDEX(mark_list, ',', 1)) + 2) AS rest
    FROM students_new

    UNION ALL

    -- Recursive member: Keep extracting marks until rest is empty
    SELECT
        student_name,
        SUBSTRING_INDEX(rest, ',', 1) AS mark,
        SUBSTRING(rest, LENGTH(SUBSTRING_INDEX(rest, ',', 1)) + 2) AS rest
    FROM split_marks
    WHERE rest != ''
)

-- Final selection of student names and their individual marks
SELECT student_name, mark, rest
FROM split_marks
WHERE mark != '';
```

```python
%%sql
WITH RECURSIVE split_marks AS (
    -- Anchor member: Get the first mark and the remaining marks
    SELECT
        student_name,
        SUBSTRING_INDEX(mark_list, ',', 1) AS mark,
        SUBSTRING(mark_list, LENGTH(SUBSTRING_INDEX(mark_list, ',', 1)) + 2) AS rest
    FROM students_new

    UNION ALL

    -- Recursive member: Keep extracting marks until rest is empty
    SELECT
        student_name,
        SUBSTRING_INDEX(rest, ',', 1) AS mark,
        SUBSTRING(rest, LENGTH(SUBSTRING_INDEX(rest, ',', 1)) + 2) AS rest
    FROM split_marks
    WHERE rest != ''
)
, `subject row no` AS (
    SELECT student_name, mark, 
           ROW_NUMBER() OVER (PARTITION BY student_name) AS `row num`
    FROM split_marks
)
, `subject rank` AS (
    SELECT student_name, mark, `row num`,
           RANK() OVER (PARTITION BY `row num` ORDER BY mark DESC) AS `rank`
    FROM `subject row no`
)
SELECT * FROM `subject rank`;
```

```python
%%sql
with rewards_split as
(
select club_id,member_id,rewards,
unnest(string_to_array(rewards, ':') ) as "new_rewards"
from rewards_table
)
, distinct_clubid_and_rewards as
(
select distinct(club_id) as club_id, new_rewards
from rewards_split
order by club_id,new_rewards
)
select club_id, new_rewards,
case when new_rewards='AA' then 10
     when new_rewards='BB' then 20
     when new_rewards='CC' then 30
     when new_rewards='DD' then 40
     when new_rewards='PP' then 50
     else 0
end as "points"
from distinct_clubid_and_rewards;
```

```python
%%sql
with rewards_split as
(
select club_id,member_id,rewards,
unnest(string_to_array(rewards, ':') ) as "new_rewards"
from rewards_table
)
, distinct_clubid_and_rewards as
(
select distinct(club_id) as club_id, new_rewards
from rewards_split
order by club_id,new_rewards
)
,rewards_with_points as
(
select club_id, new_rewards,
case when new_rewards='AA' then 10
     when new_rewards='BB' then 20
     when new_rewards='CC' then 30
     when new_rewards='DD' then 40
     when new_rewards='PP' then 50
     else 0
end as "points"
from distinct_clubid_and_rewards
)
select club_id, sum(points) as "Total points"
from rewards_with_points
group by club_id;
```
