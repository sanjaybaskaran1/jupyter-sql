--> lag(), lead(), first_value(), last_value(), nth_value(), ntile() --> value window functions 

```python
%%sql
select * from employees;
```

```python
%%sql
select *,
LAG(salary,1) OVER(ORDER BY id) as `previous_employee_salary`
from employees;
```

```python
%%sql
select *,
LAG(salary,1) OVER(ORDER BY id) as `previous_employee_salary`,
LAG(salary,1,0) OVER(ORDER BY id) as `previous_employee_salary`
from employees;
```

```python
%%sql
with `cte` as
(
SELECT *,
LAG(salary,1,0) OVER(PARTITION BY department ) as `lag_employee_salary`,
LEAD(salary,1,0) OVER(PARTITION BY department) as `lead_employee_salary`
FROM employees
)
SELECT * FROM `cte`;
```

```python
%%sql
SELECT *,
FIRST_VALUE(salary) OVER(PARTITION BY department) as `first_salary`,
LAST_VALUE(salary) OVER(PARTITION BY department) as `last_salary`,
NTH_VALUE(salary, 2) OVER(PARTITION BY department) as `nth_salary`
FROM employees;
```

```python
%%sql
SELECT *,
NTILE(2) OVER(ORDER BY id) as `2_partition`
FROM employees;
```

```python
%%sql
SELECT *,
NTILE(3) OVER(ORDER BY id) as `2_partition`
FROM employees;
```

```python
%%sql
WITH `cte` as
(
SELECT *,
NTILE(2) OVER(ORDER BY id) as `2_partition`
FROM employees
)
SELECT  `2_partition`, SUM(salary) as `Total salary`
FROM `cte`
GROUP BY `2_partition`;
```
