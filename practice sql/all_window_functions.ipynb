1) --> rank(), dense_rank(),row_number() --> ranking window functions

1) --> rank the employees based on their salary

```python
%%sql
select *,
rank() over(order by salary desc) as `rank`,
dense_rank() over(order by salary desc) as `dense_rank`
from employees;
```

2.-->find the top 1 employees based on salary

```python
%%sql
with `salary_rank` as
(
select *,
dense_rank() over(order by salary desc) as `dense_rank`
from employees
)
select * from `salary_rank`
where `dense_rank`=1;
```

3.-->find the (top 1) and (top 7th) employees based on salary

```python
%%sql
with `salary_rank` as
(
select *,
dense_rank() over(order by salary desc) as `dense_rank`
from employees
)
select * from `salary_rank`
where `dense_rank` in(1,7);
```

5.--> find the number of employees in each department

```python
%%sql
select *,
row_number() over(partition by department) as `row_number`
from employees;
```

```python
%%sql
WITH departments_rn AS (
    SELECT *,
    ROW_NUMBER() OVER (PARTITION BY department) AS `rn`
    FROM employees
)
SELECT department, max(rn) AS `no_of_employees`
FROM departments_rn
GROUP BY department;
```

2. sum() --> window function

1.--> find the (running total) and (cummulative sum) for each department

```python
%%sql
select * from employees;
```

```python
%%sql
SELECT *,
SUM(salary) OVER(PARTITION BY department ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS `running_total`
FROM employees;
```

```python
%%sql
SELECT *,
SUM(salary) OVER(PARTITION BY department ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS `running_total`,
SUM(salary) OVER(PARTITION BY department ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS `running_total`
FROM employees;
```

```python
%%sql
WITH cte AS
(
SELECT *,
SUM(salary) OVER(PARTITION BY department ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS `running_total`
FROM employees
)
SELECT department, max(`running_total`) as `Total salary`
FROM cte
GROUP BY department;
```

--> 3. avg() --> window function

(1)--> find the employees whose salary is greaterthan department average salary but lesser than the companys average salary.

```python
%%sql
SELECT *,
AVG(salary) OVER(PARTITION BY department) AS `dept_avg_salary`,
AVG(salary) OVER(order by salary ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS `company_avg_salary`
FROM employees;
```

```python
%%sql
WITH `employee_salary` AS 
(
SELECT *,
AVG(salary) OVER(PARTITION BY department) AS `dept_avg_salary`,
AVG(salary) OVER(order by salary ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS `company_avg_salary`
FROM employees
)
SELECT * FROM `employee_salary`
where (salary > `dept_avg_salary`) AND (salary < `company_avg_salary`);
```

1) -->> find the minimum and maximum salary of employees in each department 

```python
%%sql
SELECT *,
MAX(salary) OVER(PARTITION BY `department`) AS `dept_max_salary`,
MIN(salary) OVER(PARTITION BY `department`) AS `dept_min_salary`
FROM employees;
```

2) -->> find the employees whose salary is greater than 
   -->> the minimum salary of thier department but
   -->> lesser than the maximum salary of the company 

```python
%%sql
SELECT *,
MIN(salary) OVER(PARTITION BY department) AS `dept_min_salary`,
MAX(salary) OVER(ORDER BY salary ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS `company_max_salary`
FROM employees;
```

```python
%%sql
WITH `employees_salary` AS
(
SELECT *,
MIN(salary) OVER(PARTITION BY department) AS `dept_min_salary`,
MAX(salary) OVER(ORDER BY salary ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS `company_max_salary`
FROM employees
)
SELECT * FROM `employees_salary`
WHERE (salary > `dept_min_salary`) AND (salary < `company_max_salary`);
```

-->> first_value(), last_value, nth_value, lag(), lead() -->> value window functions

1) -->> lag(), lead() -->>window functions

```python
%%sql
select * from employees;
```

(1) --> find the next name, previous name of each employees
    --> if their is (no pervious name) or (next name) use (no name) as default value

```python
%%sql
SELECT *,
LAG(name,1) OVER(ORDER BY id) AS `previous_name`,
LEAD(name,1) OVER(ORDER BY id) AS `next_name`
FROM employees;
```

```python
%%sql
SELECT *,
LAG(name,1) OVER(ORDER BY id) AS `previous_name`,
LAG(name,1,"no name") OVER(ORDER BY id) AS `previous_name`,
LEAD(name,1) OVER(ORDER BY id) AS `next_name`,
LEAD(name,1,"no name") OVER(ORDER BY id) AS `next_name`
FROM employees;
```

(2) --> find the next name, previous name of each employees in a department 
    --> if their is (no pervious name) or (next name) use (no name) as default value

```python
%%sql
SELECT *,
LAG(name,1,"no name") OVER(PARTITION BY department) AS `dept_emp_previous_name`,
LEAD(name,1,"no name") OVER(PARTITION BY department) AS `dept_emp_previous_name`
FROM employees;
```

(3) --> find all the employees whose salary is greater than 
    --> the salary of the next employee in their department

```python
%%sql
SELECT *,
LEAD(salary,1,0) OVER(PARTITION BY department) AS `dept_emp_next_salary`
FROM employees;
```

```python
%%sql
WITH `employees_salary` AS
(
SELECT *,
LEAD(salary,1,0) OVER(PARTITION BY department) AS `dept_emp_next_salary`
FROM employees
)
SELECT * FROM `employees_salary`
WHERE salary > `dept_emp_next_salary`;
```

2. -->> first_value(), last_value(), nth_value() -->> window functions

```python
%%sql
SELECT * FROM employees;
```

1.-->> find the first salary and last salary and second salary in each department

```python
%%sql
SELECT *,
FIRST_VALUE(salary) OVER(PARTITION BY department) AS `dept_first_salary`,
LAST_VALUE(salary) OVER(PARTITION BY department) AS `dept_last_salary`,
NTH_VALUE(salary,2) OVER(PARTITION BY department) AS `dept_second_salary`
FROM employees;
```
