----> subqueries with joins
-----> find the employee and salary details of the employees 
------> whose salary is higher than the average salary

```python
%%sql
select e.name, e.gender,e.designation, s.salary, s.salary_date, 
case when salary > (select avg(salary) from salary) then "salary above avg salary" else "salary below avg salary" end as `salary detail`
from (employees as e) inner join (salary as s)
on e.emp_id = s.emp_id 
where e.emp_id in
(
select emp_id from salary where salary >
(select avg(salary) from salary)
);
```

----> subqueries with joins
-----> find the employee and salary details of the employees 
------> whose salary is lesser than the average salary

```python
%%sql
select e.name, e.gender,e.designation, s.salary, s.salary_date, 
case when salary > (select avg(salary) from salary) then "salary above avg salary" else "salary below avg salary" end as `salary detail`
from (employees as e) inner join (salary as s)
on e.emp_id = s.emp_id 
where e.emp_id in
(
select emp_id from salary where salary <
(select avg(salary) from salary)
);
```
