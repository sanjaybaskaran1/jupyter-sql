```python
%%sql
select word,
row_number() over(order by left(word,1))as "rn"
from Combinations;
```

```python
%%sql
select word,
row_number() over(order by left(word,1))as "rn"
from Combinations
where length(word) <= 1;
```

```python
%%sql
select word,
dense_rank() over(order by left(word,1)) as "rank"
from Combinations;
```

```python
%%sql
select word,
dense_rank() over(order by left(word,1)) as "rank"
from Combinations
where length(word) > 1;
```

```python
%%sql
with cte1 as
(
select word,
row_number() over(order by left(word,1))as "rn"
from Combinations
where length(word) <= 1
)
, cte2 as
(
select word,
dense_rank() over(order by left(word,1)) as "rank"
from Combinations
where length(word) > 1
)
select *
from cte1 inner join cte2
on cte1.rn = cte2.rank;
```

```python
%%sql
with cte1 as
(
select word,
row_number() over(order by left(word,1))as "rn"
from Combinations
where length(word) <= 1
)
, cte2 as
(
select word,
dense_rank() over(order by left(word,1)) as "rank"
from Combinations
where length(word) > 1
)
select cte1.word as "Alphabet", cte2.word 
from cte1 inner join cte2
on cte1.rn = cte2.rank;
```
