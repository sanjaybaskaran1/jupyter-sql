```python
%%sql
select location,
timestampdiff(hour,signup_start_date,signup_stop_date ) as " signup_duration"
from signups;
```

```python
%%sql
select location,
avg(timestampdiff(hour,signup_start_date,signup_stop_date )) as "average_signup_duration"
from signups
group by location;
```

```python
%%sql
select location,
avg(timestampdiff(hour,signup_start_date,signup_stop_date )) as "average_signup_duration",
round(avg(t.amt), 2) as "average_amount"
from (signups as s) inner join (transactions_1 as t)
on s.signup_id = t.signup_id
group by location;
```

```python
%%sql
with cte as
(
select location,
avg(timestampdiff(hour,signup_start_date,signup_stop_date )) as "average_signup_duration",
round(avg(t.amt), 2) as "average_amount"
from (signups as s) inner join (transactions_1 as t)
on s.signup_id = t.signup_id
group by location
)
select *,
round(average_amount / average_signup_duration, 2) as "profit_ratio"
from cte
order by profit_ratio desc;
```

```python
%%sql
with yearly_host_count as
(
select year(host_since) as "year",
count(id) as "host_in_current_year"
from airbnb_search_details
group by year
)
select *
from (yearly_host_count as t1) left join (yearly_host_count as t2)
on t1.year = t2.year;
```

```python
%%sql
with yearly_host_count as
(
select year(host_since) as "year",
count(id) as "host_in_current_year"
from airbnb_search_details
group by year
)
select t1.year as "current_year", t1.host_in_current_year,
t2.year as "previous_year", t2.host_in_current_year as "host_in_previous_year"
from (yearly_host_count as t1) left join (yearly_host_count as t2)
on t1.year = t2.year+1;
```

```python
%%sql
with yearly_host_count as
(
select year(host_since) as "year",
count(id) as "host_in_current_year"
from airbnb_search_details
group by year
)
select t1.year as "current_year", t1.host_in_current_year,
t2.year as "previous_year", t2.host_in_current_year as "host_in_previous_year"
from (yearly_host_count as t1) left join (yearly_host_count as t2)
on t1.year = t2.year+1;
```

```python
%%sql
with yearly_host_count as
(
select year(host_since) as "year",
count(id) as "host_in_current_year"
from airbnb_search_details
group by year
)
, cte2 as
(
select t1.year, t1.host_in_current_year, t2.host_in_current_year as "host_in_previous_year"
from (yearly_host_count as t1) left join (yearly_host_count as t2)
on t1.year = t2.year+1
)
select year,host_in_current_year,host_in_previous_year,
((host_in_current_year-host_in_previous_year)/host_in_previous_year)*100 as "growth_rate"
from cte2
order by year asc;
```

```python
%%sql
select 'male' as "gender",
avg(abs(gun_time - net_time)) as "avg time difference"
from marathon_male;
```

```python
%%sql
select 'female' as "gender",
avg(abs(gun_time - net_time)) as "avg time difference"
from marathon_female;
```

```python
%%sql
with cte as
(
select 'male' as "gender",
avg(abs(gun_time - net_time)) as "avg time difference"
from marathon_male

union

select 'female' as "gender",
avg(abs(gun_time - net_time)) as "avg time difference"
from marathon_female
)
select * from cte;
```

```python
%%sql
with cte as
(
select 'male' as `gender`,
avg(abs(gun_time - net_time)) as `avg_time_diff` 
from marathon_male
union
select 'female' as `gender`,
avg(abs(gun_time - net_time)) as `avg_time_diff` 
from marathon_female
)
select 
abs((select avg_time_diff from cte where gender = 'male') -
   (select avg_time_diff from cte where gender = 'female')) as `absolute time difference between male & female`;
```

```python
%%sql
select quarter(sa.sales_date) as "Quarter", sum(sa.sales_amount * sr.exchange_rate) as "Total Sales USD"
from (sf_exchange_rate as sr) inner join (sf_sales_amount as sa)
on (sr.date = sa.sales_date) AND (sa.currency = sr.source_currency AND sr.target_currency = 'USD') 
where quarter(sa.sales_date) in(1,2)
group by Quarter;
```

```python
%%sql
select quarter(sa.sales_date) as "Quarter", sum(sa.sales_amount * sr.exchange_rate) as "Total Sales USD"
from (sf_exchange_rate as sr) inner join (sf_sales_amount as sa)
on (sr.date = sa.sales_date) 
where (sa.currency = sr.source_currency = sr.target_currency in('USD')) AND quarter(sa.sales_date) in(1,2)
group by Quarter;
```

```python
%%sql
with `Country_Active_Users` as
(
select country, count(user_id) as "active users" 
from penetration_analysis
where last_active_date >= date_add('2024-01-01',interval -30 day)
AND listening_hours	>= 10 AND sessions >=5
group by country
)
select country, count(user_id) as "total users"
from penetration_analysis
group by country;
```

```python
%%sql
with `Country_Active_Users` as
(
select country, count(user_id) as "active users" 
from penetration_analysis
where last_active_date >= date_add('2024-01-01',interval -30 day)
AND listening_hours	>= 10 AND sessions >=5
group by country
)
, `Country_Total_Users` as
(
select country, count(user_id) as "total users"
from penetration_analysis
group by country
)
select *
from (Country_total_users as `tu`) inner join (Country_active_users as `au`)
on tu.country = au.country;
```

```python
%%sql
with `Country_Active_Users` as
(
select country, count(user_id) as "active users" 
from penetration_analysis
where last_active_date >= date_add('2024-01-01',interval -30 day)
AND listening_hours	>= 10 AND sessions >=5
group by country
)
, `Country_Total_Users` as
(
select country, count(user_id) as "total users"
from penetration_analysis
group by country
)
select au.country,
round((`active users`/`total users`)*100, 2) as "Active Users Penetration Rate"
from (Country_total_users as `tu`) inner join (Country_active_users as `au`)
on tu.country = au.country;
```
