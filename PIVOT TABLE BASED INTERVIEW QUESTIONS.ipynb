```python
%%sql
select * from occupation;
```

```python
%%sql
select *,
row_number() over(partition by occup order by name) as "rn"
from occupation;
```

```python
%%sql
with cte as(
select *,
row_number() over(partition by occup order by name) as "rn"
from occupation
)
select rn,
case when occup = 'singer' then name end as 'singer',
case when occup = 'professor' then name end as 'professor',
case when occup = 'doctor' then name end as 'doctor',
case when occup = 'actor' then name end as 'actor'
from cte;
```

```python
%%sql
with cte as(
select *,
row_number() over(partition by occup order by name) as "rn"
from occupation
)
select rn,
max(case when occup = 'singer' then name end) as 'singer',
max(case when occup = 'professor' then name end) as 'professor',
max(case when occup = 'doctor' then name  end) as 'doctor',
max(case when occup = 'actor' then name end) as 'actor'
from cte
group by rn;
```

```python
%%sql
select days ,
unnest(string_to_array(days, ',')) as "day_split"
from Combination;
```

```python
%%sql
with combination_new as
(
select days ,
unnest(string_to_array(days, ',')) as "day_split"
from Combination
)
select * 
from combination_new as c inner join Weeks as w
on c.day_split = w.letter;
```

```python
%%sql
with combination_new as
(
select days ,
unnest(string_to_array(days, ',')) as "day_split"
from Combination
)
select c.days, w.day 
from combination_new as c inner join Weeks as w
on c.day_split = w.letter;
```

```python
%%sql
with combination_new as
(
select days ,
unnest(string_to_array(days, ',')) as "day_split"
from Combination
)
, cte2 as
(
select c.days, w.day 
from combination_new as c inner join Weeks as w
on c.day_split = w.letter
)
select days,
max(case when day = 'Mon' then 'True' else 'Null' end) as "Mon",
max(case when day = 'Tue' then 'True' else 'Null' end) as "Tue",
max(case when day = 'Wed' then 'True' else 'Null' end) as "Wed",
max(case when day = 'Thu' then 'True' else 'Null' end) as "Thu",
max(case when day = 'Fri' then 'True' else 'Null' end) as "Fri",
max(case when day = 'Sat' then 'True' else 'Null' end) as "Sat",
max(case when day = 'Sun' then 'True' else 'Null' end) as "Sun"
from cte2
group by days
order by days asc;
```

```python
%%sql
select distinct(month) as "month",
case when category = 'Clothing' then amount end as "clothing",
case when category = 'Electronics' then amount end as "Electronics"
from sales_data;
```

```python
%%sql
select distinct(month) as "month",
max(case when category = 'Clothing' then amount end) as "clothing",
max(case when category = 'Electronics' then amount end) as "Electronics"
from sales_data
group by month
order by 2 desc;
```
